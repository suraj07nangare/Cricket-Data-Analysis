```python
batsmen = pd.read_csv('sharma-kohli.csv')
batsmen
```

```python
plt.plot(batsmen['index'],batsmen['V Kohli'])
```

```python
plt.plot(batsmen['index'],batsmen['V Kohli'], color='red',linewidth=3, label = 'virat',marker='o')
plt.plot(batsmen['index'],batsmen['RG Sharma'],color='blue',linewidth=3,label = 'rohit',marker='o')

plt.legend()
```

```python
plt.plot(batsmen['index'],batsmen['V Kohli'], color='red',linewidth=3, label = 'virat',marker='o',linestyle='dotted')
plt.plot(batsmen['index'],batsmen['RG Sharma'],color='blue',linewidth=3,label = 'rohit',marker='o',linestyle='dashed')
plt.legend(loc='upper right')

plt.title('Rohit vs Virat IPL runs')
plt.xlabel('Seasons')
plt.ylabel('runs scored')
```

```python
# bar plot on dataset
df = pd.read_csv('batsman_season_record.csv')
df.head()
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width= 0.2,color='yellow',label='2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width= 0.2,color='red',label='2016')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width= 0.2,color='green',label='2017')

plt.legend()
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='pink')
plt.xticks(rotation= 'vertical')
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```
